```python
import os
import pandas as pd

from nltk.cluster import KMeansClusterer, cosine_distance
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from utils import TextNormalizer, OneHotVectorizer
```

```python
df = pd.read_csv('../../../data/headlines/headlines_subset.csv')
```

```python
sentences = df['headline_text'].tolist()
```

```python
pipeline = Pipeline([
    ('normalizer', TextNormalizer()),
    ('vectorizer', OneHotVectorizer()),
])
X = pipeline.fit_transform(sentences)
```

```python
cluster = KMeans()
clusters = cluster.fit_transform(X)
outputs = cluster.predict(X)
```

```python
cluster = KMeansClusterer(8, cosine_distance, avoid_empty_clusters=True)
clusters = cluster.cluster(X)
# outputs = cluster.classification_probdist(X.T)
```

```python
for index, cluster in enumerate(clusters):
    print(f"Cluster {cluster}")
```
